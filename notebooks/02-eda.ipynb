```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("../data/raw/device_telemetry.csv")

df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure()
plt.hist(df["battery_health"], bins=30)
plt.title("Battery Health Distribution")
plt.xlabel("Battery Health")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
high_risk = df[df["battery_health"] < 70]
print("High risk devices:", len(high_risk))

high_risk.head()
```
