```python
import pandas as pd
import numpy as np
from pathlib import Path

np.random.seed(42)

n = 5000
df = pd.DataFrame({
    "device_id": np.arange(1, n+1),
    "battery_cycles": np.random.randint(50, 1000, n),
    "avg_temp": np.random.normal(35, 5, n).clip(20, 55),
    "screen_on_time": np.random.normal(6, 2, n).clip(0.5, 14),
    "fast_charging_count": np.random.randint(0, 300, n),
    "cpu_usage": np.random.normal(40, 10, n).clip(5, 95),
})

df["battery_health"] = (
    100
    - 0.02 * df["battery_cycles"]
    - 0.03 * df["fast_charging_count"]
    - 0.7 * (df["avg_temp"] - 30)
    - 0.4 * (df["cpu_usage"] - 35)
    + np.random.normal(0, 2, n)
).clip(40, 100)

Path("../data/raw").mkdir(parents=True, exist_ok=True)
df.to_csv("../data/raw/device_telemetry.csv", index=False)

df.head()
```
